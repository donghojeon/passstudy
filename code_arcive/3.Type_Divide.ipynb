# LSH가 완료된 데이터 type별 분류 작업

```python
def TypeDivider(df, char):
    
    start_time = time.time()
    
    print(char ,'data is loaded / length = ',len(df))
    df = df.drop(['Domain','PW', 'index', 'HashTag'], axis = 1)
    
    # Level1 = num || up || low || spc
    df['low'] = df['PWset'].apply(LowSelect)
    df['num'] = df['PWset'].apply(NumSelect)
    df['up']  = df['PWset'].apply(UpSelect)
    df['spc'] = df['PWset'].apply(SpcSelect)
    print(char, 'level 1 done')

    # Level2 =  num&low  , num&up , num&spc, low&up , low&spc, up&spc
    df['low&num'] = df['PWset'].apply(LowNumSelect)
    df['up&low']  = df['PWset'].apply(UpLowSelect)
    df['up&num']  = df['PWset'].apply(UpNumSelect)
    df['spc&low'] = df['PWset'].apply(SpcLowSelect)
    df['spc&num'] = df['PWset'].apply(SpcNumSelect)
    df['spc&up']  = df['PWset'].apply(SpcUpSelect)
    print(char, 'level 2 done')
    
    # Level3 = num&low&up, num&low&spc, num&up&spc ,low&up&spc
    df['up&num&low']  = df['PWset'].apply(UpNumLowSelect)
    df['spc&num&low'] = df['PWset'].apply(SpcNumLowSelect)
    df['spc&up&low']  = df['PWset'].apply(SpcUpLowSelect)
    df['spc&up&num']  = df['PWset'].apply(SpcUpNumSelect)
    print(char, 'level 3 done')
    
    # Level4 = All
    df['all'] = df['PWset'].apply(AllSelect)
    print(char, 'level 4 done')
    
    print('total consumed time=', (time.time()-start_time))
    
    save_path = './Preprocessed/' +str(char)+'.pickle'
    df.to_pickle(save_path)
    
    print(char ,'Data Saved')
    print('')
```
